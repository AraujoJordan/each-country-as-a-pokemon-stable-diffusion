# Explore the fine-tuning data

Examples of the image-text pairs the model was fine-tuned on. More info at: https://huggingface.co/datasets/lambdalabs/pokemon-blip-captions

```python
from datasets import load_dataset


ds = load_dataset("lambdalabs/pokemon-blip-captions", split="train")
```

```python
for i, sample in enumerate(ds):
    if i > 2:
        break
    display(sample["image"].resize((90, 90)))
    print(sample["text"])
```
